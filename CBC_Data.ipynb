```python
import numpy as np
import pandas as pd
```

```python
n = 10
```

```python
data = {'Gender' : np.random.choice(['Male', 'Famale'], size= n),
       'Age' : np.random.randint(0, 85, size= n)}

data['RBC'] = np.where(data['Age'] >= 65,
                  round(np.random.uniform(3.9, 5.6), 2),
                       np.where((data['Age'] >= 12) & (data['Age'] < 65),
                                np.where(data['Gender'] == 'Male',
                                         round(np.random.uniform(4.5, 5.9), 2),
                                         round(np.random.uniform(3.8, 5.0), 2)), np.where((data['Age'] < 12) & (data['Age'] >= 1), round(np.random.uniform(4.1, 5.5), 2),
                                                                                                                                    round(np.random.uniform(4.1, 6.1), 2))))

data['WBC'] = np.where(data['Age'] >= 65,
                  round(np.random.uniform(3.8, 10.0), 2),
                       np.where((data['Age'] >= 12) & (data['Age'] < 65),
                                np.where(data['Gender'] == 'Male',
                                         round(np.random.uniform(4.5, 11.0), 2),
                                         round(np.random.uniform(4.5, 11.0), 2)), np.where((data['Age'] < 12) & (data['Age'] >= 1), round(np.random.uniform(5, 14.5), 2),
                                                                                                                                    round(np.random.uniform(9, 30.0), 2))))

data['Hb'] = np.where(data['Age'] >= 65,
                  round(np.random.uniform(11.5, 16.0), 2),
                       np.where((data['Age'] >= 12) & (data['Age'] < 65),
                                np.where(data['Gender'] == 'Male',
                                         round(np.random.uniform(13.5, 17.5), 2),
                                         round(np.random.uniform(12, 15.5), 2)), np.where((data['Age'] < 12) & (data['Age'] >= 1), round(np.random.uniform(11.5, 15.5), 2),
                                                                                                                                    round(np.random.uniform(13.5, 23.5), 2))))
data['Hct'] = np.where(data['Age'] >= 65,
                  np.random.randint(36, 47, size= 10),
                       np.where((data['Age'] >= 12) & (data['Age'] < 65),
                                np.where(data['Gender'] == 'Male',
                                         np.random.randint(40, 50, size= 10),
                                         np.random.randint(34, 44, size= 10)), np.where((data['Age'] < 12) & (data['Age'] >= 1), np.random.randint(35, 45, size= 10),
                                                                                                                                    np.random.randint(45, 65, size= 10))))
data['PLT'] = np.where(data['Age'] >= 65,
                  np.random.randint(150, 400, size= 10),
                       np.where((data['Age'] >= 12) & (data['Age'] < 65),
                                np.where(data['Gender'] == 'Male',
                                         np.random.randint(150, 400, size= 10),
                                         np.random.randint(150, 400, size= 10)), np.where((data['Age'] < 12) & (data['Age'] >= 1), np.random.randint(150, 400, size= 10),
                                                                                                                                    np.random.randint(150, 400, size= 10))))

df = pd.DataFrame(data)
df
```
